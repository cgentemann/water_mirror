```python
import pandas as pd
import ephem
```

```python
# Function to calculate sunlight for each timestamp
def calculate_sunlight(lat, lon, timestamps):
    # Initialize observer
    observer = ephem.Observer()
    observer.lat = str(lat)
    observer.lon = str(lon)

    # Initialize the Sun object
    sun = ephem.Sun()

    # Store results
    sunlight_data = []

    for timestamp in timestamps:
        observer.date = timestamp

        # Compute the position of the sun for the observer at the given time
        sun.compute(observer)

        # Altitude is in degrees (-90 to 90), where 90 is zenith and 0 is on the horizon
        altitude = sun.alt

        # Normalize sunlight (1 for zenith, 0 for horizon, and negative for below horizon)
        normalized_sunlight = max(0, altitude / ephem.degree)

        sunlight_data.append(normalized_sunlight)

    return pd.DataFrame({'timestamp': timestamps, 'sunlight_fraction': sunlight_data})

# Latitude and longitude for the given location
lat, lon = 37.7897, -122.3972

# Generate hourly timestamps from Dec 21, 2023 to Dec 21, 2024
timestamps = pd.date_range(start="2023-12-21 19:00:00", end="2024-12-21 01:00:00", freq='0.25h')

# Calculate sunlight fraction for the entire time range
sunlight_df = calculate_sunlight(lat, lon, timestamps)

# Save the result to a CSV
sunlight_df.to_csv('sunlight_data_dec2023_to_dec2024.csv', index=False)

# Display the first few rows
print(sunlight_df.head())
```

```python
sunlight_df.tail()
```

```python
sunlight_df['sunlight_fraction'] = sunlight_df['sunlight_fraction'] / sunlight_df['sunlight_fraction'].max()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 5))

plt.plot(sunlight_df['timestamp'], sunlight_df['sunlight_fraction'])
plt.title('Sunlight Fraction')
plt.xlabel('time')
plt.ylabel('sun')
plt.grid(True)
#my_dpi = 150
#plt.savefig("sturmflut_wind.png", dpi=my_dpi * 1)
plt.show()
```

```python
# Create variables to store increasing and decreasing sunlight fractions
increasing_sunlight = []
decreasing_sunlight = []

# Set initial states for startup and startdown
startup =  1.4
startdown = -1.4
scale_factor = 3.5


# Initialize previous value for comparison
previous_value = sunlight_df['sunlight_fraction'].iloc[0]

# Loop through the sunlight_fraction column to compare row by row
for current_value in sunlight_df['sunlight_fraction']:
    if current_value > previous_value:
        increasing_sunlight.append(startdown+(current_value * scale_factor))
        decreasing_sunlight.append(startup)
    elif current_value < previous_value:
        decreasing_sunlight.append(startup-(current_value * scale_factor))
        increasing_sunlight.append(startdown)
    else:
        # When the value stays the same, reset both lists to "startup" or "startdown"
        increasing_sunlight.append(0.0)
        decreasing_sunlight.append(0.0)
    
    previous_value = current_value

# Add the new columns to the dataframe
sunlight_df['increasing_sunlight'] = increasing_sunlight
sunlight_df['decreasing_sunlight'] = decreasing_sunlight
sunlight_df.head(100)
```

```python
new_df = sunlight_df.head(400).copy()

plt.figure(figsize=(15, 5))

plt.plot(new_df['timestamp'], new_df['sunlight_fraction'])
plt.title('Sunlight Fraction')
plt.xlabel('increase')
plt.ylabel('sun')
plt.grid(True)
#my_dpi = 150
#plt.savefig("sturmflut_wind.png", dpi=my_dpi * 1)
plt.show()
```
