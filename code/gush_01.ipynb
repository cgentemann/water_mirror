```python
from PIL import Image, ImageDraw
import numpy as np
```

```python
def create_bicubic_gradient_image(width, height, gradient_start_value, gradient_end_value, gradient_start_point, gradient_end_point):
    # Create an image with a white background
    image = Image.new('L', (width, height), 'black')
    pixels = image.load()
    
    # Calculate gradient step
    gradient_step = (gradient_end_value - gradient_start_value) / (gradient_end_point - gradient_start_point)
    
    for y in range(height):
        if y < gradient_start_point:
            value = gradient_start_value
        elif y > gradient_end_point:
            value = gradient_end_value
        else:
            value = gradient_start_value + gradient_step * (y - gradient_start_point)
        
        # Ensure value is within the valid range
        value = max(0, min(255, int(value)))
        
        for x in range(width):
            pixels[x, y] = value
    
    return image

width, height = 232, 280
gradient_start_value = 255  # Starting grayscale value
gradient_end_value = 0  # Ending grayscale value

for y in range(1, height):
    gradient_start_point = 0  # Starting y-coordinate
    gradient_end_point = y  # Ending y-coordinate
    
    # Create the gradient image
    image = create_bicubic_gradient_image(width, height, gradient_start_value, gradient_end_value, gradient_start_point, gradient_end_point)
    
    # Save the image
    image_filename = '../img/gradient_image_'+str(gradient_end_point)+'.png'
    image.save(image_filename)
    
    # Optionally, display the image
    #image.show()
    
    print(f'Saved gradient image to {image_filename}')


# Generate additional 255 images with decreasing gradient_start_value
for i in range(1,255):
    gradient_start_value = 255 - i
    gradient_start_point = 0  # Starting y-coordinate
    gradient_end_point = height  # Ending y-coordinate
    
    # Create the gradient image
    image = create_bicubic_gradient_image(width, height, gradient_start_value, gradient_end_value, gradient_start_point, gradient_end_point)
    
    # Save the image
    image_filename = f'../img/gradient_image_{280 + (255-gradient_start_value)}.png'
    image.save(image_filename)
    print(f'Saved gradient image to {image_filename}')
    # Optionally, display the image
    # image.show()
```
